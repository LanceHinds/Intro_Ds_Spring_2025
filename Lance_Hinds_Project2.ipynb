```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#read dataset
df = pd.read_csv('Happiness_2015_updated_cleaned.csv')

#copy dataset for data manipulation
df_new=df.copy()

#drop countries outside of the happiness rank of 10 for closer data observations
dropped_rows = df_new[df_new['Happiness Rank'] > 10].index #determine which rows to drop based on index and label
df_new.drop(dropped_rows, inplace=True) #drops rows specified in dropped_rows variable

#extract components of dataset that will be observed 
GDP = df_new['Trust (Government Corruption)']
Country = df_new['Region']

#plot GDP and Country comparison
plt.bar(Country,GDP, color='lightgreen') #Contries are on horizontal axis and GDP is on vertical axis
plt.title("Regions and Trust in Government")
plt.xlabel("Region")
plt.ylabel("Trust in Government")

plt.savefig("image1a.jpg") #save visualization
plt.show()
```

The visualization displays the total trust in government for three regions that were included in the happiness dataset.  The three regions consists of countries that were within the top 10 of the overall happiness rank.  The visualization reveals that countries in Western Europe have the most trust in their government.  This observation could be important in understanding the main influences on a country's happiness rank.

```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#read dataset
df = pd.read_csv('Happiness_2015_updated_cleaned.csv')

#copy dataset for data manipulation
df_new=df.copy()

#drop countries outside of the happiness rank of 20 for closer data observations
dropped_rows = df_new[df_new['Happiness Rank'] > 20].index #determine which rows to drop based on index and label
df_new.drop(dropped_rows, inplace=True) #drops rows specified in dropped_rows variable

plt.figure(figsize=(20,10))
sns.boxplot(data=df_new, x='Region', y='Health (Life Expectancy)')
plt.title("Score Distribution by Subject (Seaborn)")
plt.xlabel("Region")
plt.ylabel("Health (Life Expectancy)")
#plt.show()

plt.savefig("image1b.jpg") #save visualization
```

The visualization displays the range of the health factor of regions included in the dataset.  This visualization utilizes the regions of the top 20 countries by happiness rank.  The visualization reveals that most counties have a relative high positive health factor, but some countires have low factors (seen by the outliers).

```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#read dataset
df = pd.read_csv('Happiness_2015_updated_cleaned.csv')

Dev_GDP=df.groupby('Development Level')['Economy (GDP per Capita)'].mean() #average of GDP per development level
#print(Dev_GDP)
Dev_GDP.plot(kind='bar',color=['blue','red','green','orange'],edgecolor='black')
plt.title("Average GDP vs Development Level")
plt.xlabel("Development Level")
plt.ylabel("Average GDP")

plt.savefig("image2.jpg") #save visualization
```

Some differences that are observed among the different development levels are that the Average GDP follows a trend.  Countries with a higher development level has a higher GDP than countries with a lower development level.  This means that GDP is a huge factor in the develpment and happiness of a country.

```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#read dataset
df = pd.read_csv('Happiness_2015_updated_cleaned.csv')

df_new=df.copy()

#drop rows that won't be used in heatmap
df_new.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
df_new.drop(columns=['Unnamed: 0.1'], axis=1, inplace=True)

#plot heatmap
plt.figure(figsize=(10,10))
df_cor2 = df_new.corr(numeric_only=True)
sns.heatmap(df_cor2, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix (Seaborn)")

plt.savefig("image3b.jpg") #save visualization
```

The heatmap shows a positive correlation between happiness rank and freedom, meaning that countries ranked lower in the happiness rank have more freedom than countries at the top of the rank.
